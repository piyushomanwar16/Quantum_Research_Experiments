```python
import numpy as np
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

from qiskit_machine_learning.kernels import FidelityStatevectorKernel
from qiskit.circuit import QuantumCircuit, ParameterVector

# -------------------------------------------------------------
# 1. LOAD & PREPROCESS DATA (Binary: Class 0 vs Class 1)
# -------------------------------------------------------------
data = load_wine()
X = data.data
y = data.target

# Keep only class 0 and 1 (binary classification works better for QSVM)
mask = y < 2
X = X[mask]
y = y[mask]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reduce dimensionality to 2 features (QSVM works better with small dims)
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_scaled)

X_train, X_test, y_train, y_test = train_test_split(
    X_reduced, y, test_size=0.2, random_state=42
)

# -------------------------------------------------------------
# 2. CUSTOM FEATURE MAP
# -------------------------------------------------------------
def custom_feature_map(d, reps=2):
    x = ParameterVector('x', d)
    qc = QuantumCircuit(d)

    for r in range(reps):
        # Nonlinear data re-uploading
        for i in range(d):
            qc.rx(x[i], i)
            qc.rz(x[i]**2, i)

        # Entanglement
        for i in range(d - 1):
            qc.cz(i, i+1)

    return qc

feature_map = custom_feature_map(d=2, reps=3)
print("Custom Feature Map Created Successfully!")
print(feature_map)

# -------------------------------------------------------------
# 3. QUANTUM KERNEL
# -------------------------------------------------------------
quantum_kernel = FidelityStatevectorKernel(feature_map=feature_map)

# -------------------------------------------------------------
# 4. COMPUTE KERNEL MATRICES
# -------------------------------------------------------------
K_train = quantum_kernel.evaluate(X_train, X_train)
K_test = quantum_kernel.evaluate(X_test, X_train)

print("\nK_train shape:", K_train.shape)
print(K_train)

print("\nK_test shape:", K_test.shape)
print(K_test)

# -------------------------------------------------------------
# 5. TRAIN CLASSICAL SVM WITH QUANTUM KERNEL
# -------------------------------------------------------------
clf = SVC(kernel="precomputed")
clf.fit(K_train, y_train)

accuracy = clf.score(K_test, y_test)
print("\nFinal QSVM Accuracy:", accuracy)
```

```python
import numpy as np
from sklearn.metrics.pairwise import pairwise_kernels
from sklearn.svm import SVC
from qiskit_machine_learning.kernels import FidelityStatevectorKernel
from qiskit.circuit import QuantumCircuit, ParameterVector

# ---------- Alignment Function ----------
def kernel_alignment(K, y):
    y = y.reshape(-1, 1)
    yyT = y @ y.T
    return np.sum(K * yyT) / np.sqrt(np.sum(K**2) * np.sum(yyT**2))

# ---------- Custom parametrized feature map ----------
def custom_map(d, reps=2, alpha=1.0):
    x = ParameterVector("x", d)
    qc = QuantumCircuit(d)

    for r in range(reps):
        for i in range(d):
            qc.rx(alpha * x[i], i)
            qc.rz(alpha * x[i] ** 2, i)
        for i in range(d - 1):
            qc.cz(i, i+1)
    return qc

# ---------- Search space ----------
reps_list = [1, 2, 3, 4]
alpha_list = [0.5, 1.0, 1.5, 2.0]

best_acc = 0
best_params = None

for reps in reps_list:
    for alpha in alpha_list:
        feature_map = custom_map(d=2, reps=reps, alpha=alpha)
        qkernel = FidelityStatevectorKernel(feature_map=feature_map)

        K_train = qkernel.evaluate(X_train, X_train)
        K_test = qkernel.evaluate(X_test, X_train)

        clf = SVC(kernel="precomputed")
        clf.fit(K_train, y_train)
        acc = clf.score(K_test, y_test)

        print(f"Trying reps={reps}, alpha={alpha} â†’ acc={acc}")

        if acc > best_acc:
            best_acc = acc
            best_params = (reps, alpha)

print("\nðŸ”¥ Best Accuracy Found:", best_acc)
print("Best Params:", best_params)
```

```python
# --- Required imports (run once) ---
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.decomposition import KernelPCA
from sklearn.metrics.pairwise import rbf_kernel
from sklearn.model_selection import GridSearchCV
from qiskit.circuit.library import ZZFeatureMap
from qiskit_machine_learning.kernels import FidelityStatevectorKernel
from qiskit.circuit import QuantumCircuit, ParameterVector

# --- Rebuild best feature-map from your search (reps=2, alpha=0.5) ---
def custom_map(d, reps=2, alpha=0.5):
    x = ParameterVector("x", d)
    qc = QuantumCircuit(d)
    for r in range(reps):
        for i in range(d):
            qc.rx(alpha * x[i], i)
            qc.rz(alpha * x[i] ** 2, i)
        for i in range(d - 1):
            qc.cz(i, i+1)
    return qc

# If you already have X_train/X_test/y_train/y_test in memory, good.
# Otherwise, load/split your data first (not included here).

# --- Create the quantum kernel (statevector) with that feature map ---
d = X_train.shape[1]
feature_map_best = custom_map(d=d, reps=2, alpha=0.5)
qkernel = FidelityStatevectorKernel(feature_map=feature_map_best)

# --- Compute Gram matrices (train & test) ---
print("Computing quantum Gram matrices (statevector) ...")
K_train = qkernel.evaluate(X_train, X_train)   # shape (n_train, n_train)
K_test  = qkernel.evaluate(X_test,  X_train)   # shape (n_test,  n_train)
print("K_train shape:", K_train.shape, "K_test shape:", K_test.shape)

# -----------------------------
# 1) Kernel PCA -> classical SVM
# -----------------------------
print("\n== KernelPCA -> SVM (using quantum Gram) ==")
# choose number of components (try a few)
for n_comp in [10, 20, 30]:
    kpca = KernelPCA(n_components=min(n_comp, K_train.shape[0]), kernel='precomputed', eigen_solver='auto')
    Z_train = kpca.fit_transform(K_train)   # features for training
    Z_test  = kpca.transform(K_test)        # features for test (transform uses precomputed kernel)
    clf = SVC(kernel='rbf')                 # standard RBF SVM on reduced features
    clf.fit(Z_train, y_train)
    y_pred = clf.predict(Z_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"KernelPCA n_comp={n_comp}  â†’ Test accuracy = {acc:.4f}")

# ---------------------------------------
# 2) Hybrid kernel: w * K_quantum + (1-w) * K_rbf
# ---------------------------------------
print("\n== Hybrid Kernel (quantum + RBF) search ==")
# compute classical RBF Gram (using same train/test X)
gamma = 1.0 / X_train.shape[1]   # default heuristic; you can tune it
K_rbf_train = rbf_kernel(X_train, X_train, gamma=gamma)
K_rbf_test  = rbf_kernel(X_test,  X_train, gamma=gamma)

best_h_acc = 0
best_w = None
for w in np.linspace(0.0, 1.0, 11):   # 0.0..1.0 step 0.1
    K_h_train = w * K_train + (1.0 - w) * K_rbf_train
    K_h_test  = w * K_test  + (1.0 - w) * K_rbf_test

    clf = SVC(kernel='precomputed')
    clf.fit(K_h_train, y_train)
    y_pred = clf.predict(K_h_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"w={w:.2f} â†’ acc={acc:.4f}")
    if acc > best_h_acc:
        best_h_acc = acc
        best_w = w

print(f"Best hybrid weight w={best_w:.2f} â†’ acc={best_h_acc:.4f}")

# ----------------------------------------------------
# 3) Combine KernelPCA + Hybrid: do KPCA on hybrid Gram
# ----------------------------------------------------
print("\n== KernelPCA on Hybrid Kernel (best w) ==")
if best_w is None:
    best_w = 0.5
K_h_train = best_w * K_train + (1.0 - best_w) * K_rbf_train
K_h_test  = best_w * K_test  + (1.0 - best_w) * K_rbf_test

for n_comp in [10, 20, 30]:
    kpca = KernelPCA(n_components=min(n_comp, K_h_train.shape[0]), kernel='precomputed')
    Z_train = kpca.fit_transform(K_h_train)
    Z_test  = kpca.transform(K_h_test)
    clf = SVC(kernel='rbf')
    clf.fit(Z_train, y_train)
    y_pred = clf.predict(Z_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"Hybrid-KPCA n_comp={n_comp} â†’ Test accuracy = {acc:.4f}")

# -----------------------
# 4) Save the best results
# -----------------------
print("\n== Summary ==")
print("Best hybrid weight:", best_w, "Best hybrid accuracy:", best_h_acc)
```

```python
# Step 4 â€” Wine Dataset + Preprocessing

from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
import numpy as np

# ----------------------------
# 1. Load Wine dataset
# ----------------------------
wine = load_wine()
X = wine.data
y = wine.target

# ----------------------------
# 2. Convert to binary classification (class 0 vs class 1)
# ----------------------------
mask = (y == 0) | (y == 1)
X = X[mask]
y = y[mask]

print("Dataset shape:", X.shape)
print("Classes:", np.unique(y))

# ----------------------------
# 3. Standardize features
# ----------------------------
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# ----------------------------
# 4. Reduce dimension â†’ 2 features (for 2 qubits)
# ----------------------------
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_std)

# ----------------------------
# 5. Scale to [0, Ï€] for angle encoding
# ----------------------------
X_scaled = (X_pca - X_pca.min()) / (X_pca.max() - X_pca.min())
X_scaled = X_scaled * np.pi

# ----------------------------
# 6. Train-test split
# ----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y
)

print("Training samples:", X_train.shape[0])
print("Testing samples:", X_test.shape[0])
```

```python
# Step 5 â€” Classical SVM Experiments

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score
import time

results = {}

# ----------------------------
# Linear SVM
# ----------------------------
start = time.time()
clf_linear = SVC(kernel="linear")
clf_linear.fit(X_train, y_train)
train_time_linear = time.time() - start

start = time.time()
y_pred_linear = clf_linear.predict(X_test)
pred_time_linear = time.time() - start

acc_linear = accuracy_score(y_test, y_pred_linear)
f1_linear = f1_score(y_test, y_pred_linear)

results["Linear SVM"] = {
    "Accuracy": acc_linear,
    "F1 Score": f1_linear,
    "Train Time": train_time_linear,
    "Predict Time": pred_time_linear
}

# ----------------------------
# RBF SVM
# ----------------------------
start = time.time()
clf_rbf = SVC(kernel="rbf", gamma="scale")
clf_rbf.fit(X_train, y_train)
train_time_rbf = time.time() - start

start = time.time()
y_pred_rbf = clf_rbf.predict(X_test)
pred_time_rbf = time.time() - start

acc_rbf = accuracy_score(y_test, y_pred_rbf)
f1_rbf = f1_score(y_test, y_pred_rbf)

results["RBF SVM"] = {
    "Accuracy": acc_rbf,
    "F1 Score": f1_rbf,
    "Train Time": train_time_rbf,
    "Predict Time": pred_time_rbf
}

# ----------------------------
# Print results
# ----------------------------
print("=== Classical SVM Results ===")
for model, metrics in results.items():
    print(f"\n{model}:")
    for m, v in metrics.items():
        print(f"  {m}: {v:.4f}")
```

```python
# Step 6 â€” QSVM using Ideal Statevector Kernel (Qiskit 1.x compatible)

from qiskit.circuit.library import ZZFeatureMap
from qiskit.quantum_info import Statevector
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score
import numpy as np
import time

# ---------------------------------------------------
# 1. Build Feature Map
# ---------------------------------------------------
feature_dim = X_train.shape[1]
feature_map = ZZFeatureMap(feature_dimension=feature_dim, reps=1)

# ---------------------------------------------------
# Helper: Convert data point â†’ quantum statevector
# ---------------------------------------------------
def encode_state(x):
    # Convert x (numpy array) â†’ dict for parameters
    param_dict = {feature_map.parameters[i]: float(x[i]) for i in range(feature_dim)}

    # Create the circuit with assigned parameters
    circ = feature_map.assign_parameters(param_dict)

    # Return the statevector |Ï†(x)>
    return Statevector.from_instruction(circ)


# ---------------------------------------------------
# 2. Compute Quantum Kernel Matrix (Statevector)
# ---------------------------------------------------
def compute_kernel_matrix(X1, X2):
    n1, n2 = len(X1), len(X2)
    K = np.zeros((n1, n2))

    # Precompute all statevectors
    sv1 = [encode_state(x) for x in X1]
    sv2 = [encode_state(x) for x in X2]

    for i in range(n1):
        for j in range(n2):
            inner = sv1[i].inner(sv2[j])
            K[i, j] = np.abs(inner)**2   # kernel = |<Ï†(xi)|Ï†(xj)>|^2

    return K


# ---------------------------------------------------
# 3. Compute Gram Matrices
# ---------------------------------------------------
print("Computing Quantum Gram Matrix (Statevector)...")
start = time.time()

K_train = compute_kernel_matrix(X_train, X_train)
K_test  = compute_kernel_matrix(X_test, X_train)

kernel_time = time.time() - start

print("K_train shape:", K_train.shape)
print("K_test shape:", K_test.shape)

# ---------------------------------------------------
# 4. Train QSVM
# ---------------------------------------------------
clf_qsvm = SVC(kernel="precomputed")

start = time.time()
clf_qsvm.fit(K_train, y_train)
train_time_qsvm = time.time() - start

start = time.time()
y_pred_qsvm = clf_qsvm.predict(K_test)
pred_time_qsvm = time.time() - start

# ---------------------------------------------------
# 5. Metrics
# ---------------------------------------------------
acc_qsvm = accuracy_score(y_test, y_pred_qsvm)
f1_qsvm  = f1_score(y_test, y_pred_qsvm)

print("\n=== QSVM (Ideal Statevector) Results ===")
print("Accuracy:", acc_qsvm)
print("F1 Score:", f1_qsvm)
print("Kernel Computation Time:", kernel_time)
print("Train Time:", train_time_qsvm)
print("Predict Time:", pred_time_qsvm)
```

```python
# Step 7 â€” QSVM with Noisy Simulator (Depolarizing Noise)

from qiskit_aer import AerSimulator
from qiskit_aer.noise import NoiseModel, depolarizing_error
from qiskit import transpile
import numpy as np
import time

def compute_noisy_kernel(X1, X2, noise_p, shots=1024):
    n1, n2 = len(X1), len(X2)
    K = np.zeros((n1, n2))

    # Build noise model
    noise_model = NoiseModel()
    noise_model.add_all_qubit_quantum_error(depolarizing_error(noise_p, 1), ["rx","ry","rz"])
    noise_model.add_all_qubit_quantum_error(depolarizing_error(noise_p, 2), ["cz"])

    backend = AerSimulator(noise_model=noise_model, shots=shots)

    for i in range(n1):
        for j in range(n2):

            # Build circuits U(x_i) and U(x_j)
            circ_i = feature_map.assign_parameters(
                {feature_map.parameters[k]: float(X1[i][k]) for k in range(feature_dim)}
            )

            circ_j = feature_map.assign_parameters(
                {feature_map.parameters[k]: float(X2[j][k]) for k in range(feature_dim)}
            )

            # Kernel circuit = U(x_i)â€  âˆ˜ U(x_j)
            circ = circ_i.inverse().compose(circ_j)

            # Add measurement
            circ.measure_all()

            # Run circuit on noisy simulator
            tcirc = transpile(circ, backend)
            result = backend.run(tcirc).result()
            counts = result.get_counts()

            # Kernel value = probability of measuring all zeros
            K[i, j] = counts.get('0' * feature_dim, 0) / shots

    return K


# Evaluate all noise levels
noise_levels = [0.01, 0.05, 0.10]
noisy_results = []

for p in noise_levels:
    print(f"\n=== Running Noisy QSVM (p = {p}) ===")

    start = time.time()
    K_train_noisy = compute_noisy_kernel(X_train, X_train, p)
    K_test_noisy  = compute_noisy_kernel(X_test, X_train, p)
    kernel_time = time.time() - start

    clf_noisy = SVC(kernel="precomputed")
    clf_noisy.fit(K_train_noisy, y_train)
    y_pred_noisy = clf_noisy.predict(K_test_noisy)

    acc = accuracy_score(y_test, y_pred_noisy)
    f1  = f1_score(y_test, y_pred_noisy)

    print("Accuracy:", acc)
    print("F1 Score:", f1)
    print("Kernel Time:", kernel_time)

    noisy_results.append((p, acc, f1, kernel_time))
```

```python
# ====== Cell A: Installs (run if needed) ======
# Uncomment if you need to (skip if already installed and working)
# !pip install --quiet qiskit qiskit-aer qiskit-machine-learning matplotlib seaborn pandas scikit-learn

# ====== Cell B: Imports & dataset prep ======
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score

from qiskit.circuit.library import ZZFeatureMap, ZFeatureMap
from qiskit import QuantumCircuit
from qiskit_machine_learning.kernels import FidelityStatevectorKernel

# Fix plotting style
sns.set(style="whitegrid")

# Load & prepare wine dataset (the same processing you used)
data = load_wine()
X = data.data
y = data.target

# If your task is binary and you've encoded labels earlier,
# ensure y has only two classes; otherwise the code below works with multi-class SVM too.
# If you earlier used binary selection, keep that. Here I assume binary for consistency:
# (If your y already binary 0/1, skip the filtering.)
if len(np.unique(y)) > 2:
    # Example: make binary by mapping class 0 -> 0, classes 1&2 -> 1 (only do if that's what you used)
    y = (y != 0).astype(int)

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train/test split (use same split you used earlier)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print("Dataset shape:", X.shape)
print("Train/test:", X_train.shape[0], "/", X_test.shape[0])
print("Classes:", np.unique(y))

# ====== Cell C: Define feature maps (3 maps) ======
feature_dim = X_train.shape[1]

# 1) ZZ feature map (entangling)
zz = ZZFeatureMap(feature_dimension=feature_dim, reps=1)

# 2) Z feature map (single-qubit Z rotations)
zmap = ZFeatureMap(feature_dimension=feature_dim, reps=1)

# 3) Custom small feature map (RX + RZ + nearest CZ)
def custom_feature_map(d, reps=1):
    x = QuantumCircuit(d)
    for r in range(reps):
        for i in range(d):
            x.rx(np.pi * (i+1) * 0.5, i)  # placeholder param-free rotation pattern (we will bind later)
            x.rz(np.pi * (i+1) * 0.25, i)
        for i in range(d-1):
            x.cz(i, i+1)
    return x

# We'll use a parameterized custom map: to use with the kernel we need Parameters,
# so build a parameterized circuit similar to your earlier custom_feature_map (but small)
from qiskit.circuit import ParameterVector, QuantumCircuit
def custom_param_feature_map(d, reps=1):
    x = ParameterVector("x", d)
    qc = QuantumCircuit(d)
    for r in range(reps):
        for i in range(d):
            qc.rx(x[i], i)
            qc.rz(x[i]**2, i)
        for i in range(d-1):
            qc.cz(i, i+1)
    return qc

custom = custom_param_feature_map(feature_dim, reps=1)

feature_maps = {
    "ZZFeatureMap": zz,
    "ZFeatureMap": zmap,
    "CustomParamMap": custom
}

# ====== Cell D: Runner - compute statevector kernels, train SVMs, collect metrics ======
results = []
kernels = {}      # store K_train and K_test for plotting later
timings = {}

for name, fmap in feature_maps.items():
    print(f"\n>> Processing feature map: {name}")

    start_k = time.time()
    # Create statevector fidelity kernel
    qkernel = FidelityStatevectorKernel(feature_map=fmap)

    # Compute Gram matrices
    K_train = qkernel.evaluate(X_train, X_train)   # shape (n_train, n_train)
    K_test  = qkernel.evaluate(X_test, X_train)    # shape (n_test, n_train)
    ktime = time.time() - start_k

    kernels[name] = (K_train, K_test)
    timings[name] = ktime

    # Train classical SVM with precomputed kernel
    clf = SVC(kernel="precomputed")
    t0 = time.time()
    clf.fit(K_train, y_train)
    train_time = time.time() - t0

    # Predict
    t1 = time.time()
    y_pred = clf.predict(K_test)
    predict_time = time.time() - t1

    acc = accuracy_score(y_test, y_pred)
    f1  = f1_score(y_test, y_pred, average='weighted')  # use weighted for multiclass / binary consistency

    results.append({
        "feature_map": name,
        "accuracy": acc,
        "f1_score": f1,
        "kernel_time_s": ktime,
        "svm_train_s": train_time,
        "svm_predict_s": predict_time,
        "n_train": K_train.shape[0],
        "n_test": K_test.shape[0]
    })

    print(f"  Acc={acc:.4f}  F1={f1:.4f}  kernel_time={ktime:.3f}s  train={train_time:.4f}s predict={predict_time:.4f}s")

# Build DataFrame
df_results = pd.DataFrame(results).set_index("feature_map")
print("\n=== Summary Table ===")
display(df_results)

# Save results CSV
df_results.to_csv("qsvm_featuremap_comparison_results.csv")

# ====== Cell E: Heatmaps of train Gram matrices ======
# Plot heatmaps side-by-side
n_maps = len(kernels)
fig, axes = plt.subplots(1, n_maps, figsize=(5 * n_maps, 4))

if n_maps == 1:
    axes = [axes]

for ax, (name, (K_train, K_test)) in zip(axes, kernels.items()):
    sns.heatmap(K_train, ax=ax, cmap="viridis", vmin=0, vmax=1)
    ax.set_title(f"{name}\nK_train heatmap")
    ax.set_xlabel("Train index")
    ax.set_ylabel("Train index")

plt.tight_layout()
plt.savefig("kernels_heatmaps.png", dpi=200)
plt.show()

# ====== Cell F: Bar charts - Accuracy and Kernel time ======
fig, ax = plt.subplots(1, 2, figsize=(12,4))

# Accuracy bar
df_results['accuracy'].plot(kind='bar', ax=ax[0], rot=0)
ax[0].set_ylim(0,1)
ax[0].set_title("QSVM Accuracy by Feature Map")
ax[0].set_ylabel("Accuracy")

# Kernel time bar
df_results['kernel_time_s'].plot(kind='bar', ax=ax[1], rot=0, color='orange')
ax[1].set_title("Kernel Computation Time (s) by Feature Map")
ax[1].set_ylabel("Seconds")

plt.tight_layout()
plt.savefig("qsvm_accuracy_kernel_time.png", dpi=200)
plt.show()

# ====== Cell G: Save kernel matrices (optional) ======
# Save first kernel as CSV (if you want)
for name, (K_train, K_test) in kernels.items():
    np.savetxt(f"K_train_{name}.csv", K_train, delimiter=",")
    np.savetxt(f"K_test_{name}.csv", K_test, delimiter=",")

print("Saved CSVs and PNGs in working directory.")
```
